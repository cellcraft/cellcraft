```python
df = pd.read_csv('../x3d/test6.csv')
df.pdbid = df.pdbid.str.replace('\'', '')
df.pdbid = df.pdbid.str.upper()
df2 = df[~df.Gosom.isnull()]
df2
```

```python
GObioproc = {}
GOmolefunt = {}
GOcellcomp = {}

df2 = df[~df.Gosom.isnull()]
for i in range(df2.shape[0]):
    row = df2.iloc[i]
    Protcomplex = protein_complex(row.pdbid)
    Protcomplex.get_PDB()
    Protcomplex.clean_pdb()
    Protcomplex.split_complex()
    Protcomplex.load_chain_info()
    

    for p in Protcomplex.proteins:
        for func in p.GObioproc:
            if func in GObioproc:
                GObioproc[func] += row.nItems
            else:
                GObioproc[func] = row.nItems
        for func in p.GOmolefunt:
            if func in GOmolefunt:
                GOmolefunt[func] += row.nItems
            else:
                GOmolefunt[func] = row.nItems
        for func in p.GOcellcomp:
            if func in GOcellcomp:
                GOcellcomp[func] += row.nItems
            else:
                GOcellcomp[func] = row.nItems
```

```python
print(GObioproc)
print(GOmolefunt)
print(GOcellcomp)
pd.DataFrame.from_dict(GObioproc,orient='index').to_csv('GObioproc.csv')
pd.DataFrame.from_dict(GOmolefunt,orient='index').to_csv('GOmolefunt.csv')
pd.DataFrame.from_dict(GOcellcomp,orient='index').to_csv('GOcellcomp.csv')
```

```python
names = ['cellular component','shit1','shit2','molecular function','shit3','shit4','biological processes']

d = {'name':'GOterms',
    'children':[]}
for dic,name in zip([GOcellcomp,{},{},GOmolefunt,{},{},GObioproc],names):
    gd = {'name':name,
        'children':[{'name':name,
        'children':[]}]}
    for term in dic.items():
        tg = {'name':term[0],'size':int(term[1])}
        gd['children'][0]['children'].append(tg)
    d['children'].append(gd)

print(d)
import json
with open('HIV.json', 'w') as fp:
    json.dump(d, fp)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import matplotlib

mdata = pd.read_csv('/home/lbrinkm/Downloads/timeline.csv')
mdata = mdata[~(mdata.year > 2015)]
j1 = mdata.loc[~mdata['year'].isnull(),['year','percentage']]
j2 = mdata.loc[~mdata['year.1'].isnull(),['year.1','percentage.1']]
j3 = mdata.loc[~mdata['year.2'].isnull(),['year.2','percentage.2']]
j4 = mdata.loc[~mdata['year.3'].isnull(),['year.3','percentage.3']]

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
plt.plot(j1['year'].tolist(),j1['percentage'].tolist()/j1['percentage'].tolist()[11],lw=3)
plt.plot(j2['year.1'].tolist(),j2['percentage.1'].tolist()/j2['percentage.1'].tolist()[11],lw=3)
plt.plot(j3['year.2'].tolist(),j3['percentage.2'].tolist()/j3['percentage.2'].tolist()[11],lw=3)
plt.plot(j4['year.3'].tolist(),j4['percentage.3'].tolist()/j4['percentage.3'].tolist()[11],lw=3)

plt.xlim(1970,2018)

plt.legend(['bioinformatics', 'biophysics', 'computational biology', 'cell biology'], loc='upper left')
plt.savefig("Graph2.pdf")
plt.clf()
```
